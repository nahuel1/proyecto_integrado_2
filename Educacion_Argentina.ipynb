```r
# prompt: Solicita a prompt Gemeni que armae un dataframe de al menos 100 filas con datos random con apellido y nombre, nro_dni, materias de educacion nivel secundario de la argentina que contenga las siguientes columnas: materia escolar, nota 1er Trimestre, nota 2do Trimestre, nota 3er Trimestre, ausencias a clases.

import pandas as pd
import random
import numpy as np

# Lista de apellidos y nombres comunes en Argentina
apellidos = ["García", "González", "Rodríguez", "Fernández", "López", "Martínez", "Pérez", "Gómez", "Díaz", "Sánchez"]
nombres = ["Juan", "María", "José", "Laura", "Carlos", "Ana", "Pedro", "Sofía", "Diego", "Martina"]

# Lista de materias de educación secundaria en Argentina
materias = ["Matemática", "Lengua y Literatura", "Historia", "Geografía", "Ciencias Naturales", "Física", "Química", "Biología", "Educación Física", "Inglés", "Arte", "Música"]

# Crear una lista de datos vacía
data = []

# Generar 100 filas de datos aleatorios
for i in range(100):
    apellido = random.choice(apellidos)
    nombre = random.choice(nombres)
    dni = random.randint(10000000, 99999999)
    materia = random.choice(materias)
    nota_1er_trimestre = round(random.uniform(1, 10), 2)
    nota_2do_trimestre = round(random.uniform(1, 10), 2)
    nota_3er_trimestre = round(random.uniform(1, 10), 2)
    ausencias = random.randint(0, 15)

    data.append([apellido, nombre, dni, materia, nota_1er_trimestre, nota_2do_trimestre, nota_3er_trimestre, ausencias])

# Crear el DataFrame de Pandas
df = pd.DataFrame(data, columns=['Apellido', 'Nombre', 'Nro_DNI', 'Materia Escolar', 'Nota 1er Trimestre', 'Nota 2do Trimestre', 'Nota 3er Trimestre', 'Ausencias a Clases'])

# Mostrar el DataFrame
df
```

```r
# prompt: Mediante el marco de datos df: generar un boton para descargar el archivo con el nombre datos_alumnos en formato CSV

import ipywidgets as widgets
from IPython.display import display
from io import StringIO
import pandas as pd

# Crear un botón para descargar el DataFrame en formato CSV
def download_csv(dataframe, filename):
    csv_data = dataframe.to_csv(index=False)
    link = widgets.HTML(
        value="<a href='data:text/csv;charset=utf-8,%s' download='%s.csv'>Descargar CSV</a>" % (
            csv_data.replace("\n", "%0A"), filename
        )
    )
    display(link)

# Crear el botón y la función para descargar
button_download = widgets.Button(description="Descargar Datos")
def on_button_clicked(b):
    download_csv(df, "datos_alumnos")
button_download.on_click(on_button_clicked)
display(button_download)
```

```r
# prompt: Mediante el marco de datos df: Armar grafico de barras con Promedio general de cada materia, ordenados de mayor a menor, agregar las faltas como valor constrastable en dos ejes. Agregar las leyendas que ayuden a la interpretación.

import pandas as pd
import matplotlib.pyplot as plt

# Calcular el promedio general de cada materia
promedio_materias = df.groupby('Materia Escolar')[['Nota 1er Trimestre', 'Nota 2do Trimestre', 'Nota 3er Trimestre']].mean().mean(axis=1)

# Ordenar los promedios de mayor a menor
promedio_materias_ordenado = promedio_materias.sort_values(ascending=False)

# Calcular la cantidad de faltas por materia
faltas_por_materia = df.groupby('Materia Escolar')['Ausencias a Clases'].sum()

# Crear el gráfico de barras con el promedio de cada materia
plt.figure(figsize=(12, 6))
ax1 = plt.subplot(111)
promedio_materias_ordenado.plot(kind='bar', color='skyblue', ax=ax1)

# Configurar el título y las etiquetas de los ejes
plt.title('Promedio General de Notas por Materia', fontsize=16)
plt.xlabel('Materia Escolar', fontsize=14)
plt.ylabel('Promedio General', fontsize=14)

# Agregar las faltas como un segundo eje
ax2 = ax1.twinx()
faltas_por_materia.plot(kind='line', color='red', marker='o', ax=ax2)
ax2.set_ylabel('Cantidad de Faltas', fontsize=14)

# Configurar la leyenda
plt.legend(['Promedio General', 'Faltas'], loc='upper right')

# Rotar las etiquetas del eje x para mejor legibilidad
plt.xticks(rotation=45, ha='right')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
